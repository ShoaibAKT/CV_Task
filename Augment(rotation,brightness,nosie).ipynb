```python
import os
import cv2
import numpy as np
from skimage.util import random_noise

# Task 3: Augmenting Images

# i) Rotation
def augment_rotation(image_path, output_dir, num_rotations=72):
    image_name = os.path.splitext(os.path.basename(image_path))[0]
    image = cv2.imread(image_path)
    height, width = image.shape[:2]
    
    for angle in range(0, 360, 5):
        rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)
        rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
        
        output_name = f"{image_name}_{angle}degree.jpeg"
        output_path = os.path.join(output_dir, output_name)
        cv2.imwrite(output_path, rotated_image)

# ii) Brightness
def augment_brightness(image_path, output_dir, num_images=5):
    image_name = os.path.splitext(os.path.basename(image_path))[0]
    image = cv2.imread(image_path)
    
    for i in range(num_images):
        brightness_factor = np.random.uniform(0.5, 1.5)
        brightened_image = cv2.convertScaleAbs(image, alpha=brightness_factor)
        
        output_name = f"{image_name}_brightness{i+1}.jpeg"
        output_path = os.path.join(output_dir, output_name)
        cv2.imwrite(output_path, brightened_image)

# iii) Noise
def augment_noise(image_path, output_dir, num_images=5):
    image_name = os.path.splitext(os.path.basename(image_path))[0]
    image = cv2.imread(image_path)
    
    for i in range(num_images):
        noisy_image = random_noise(image, mode='gaussian', var=0.01)
        noisy_image = np.array(255 * noisy_image, dtype=np.uint8)
        
        output_name = f"{image_name}_noise{i+1}.jpeg"
        output_path = os.path.join(output_dir, output_name)
        cv2.imwrite(output_path, noisy_image)


# Provide the paths and directories
input_dir = 'C:\\Users\\Shoaib\\Downloads\\cat_dog'
output_dir_rotation = 'C:\\Users\\Shoaib\\Downloads\\augmented_images_rotation'
output_dir_brightness = 'C:\\Users\\Shoaib\\Downloads\\augmented_images_brightness'
output_dir_noise = 'C:\\Users\\Shoaib\\Downloads\\augmented_images_noise'

# Create output directories if they don't exist
os.makedirs(output_dir_rotation, exist_ok=True)
os.makedirs(output_dir_brightness, exist_ok=True)
os.makedirs(output_dir_noise, exist_ok=True)

# Iterate over the images in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".jpeg"):
        image_path = os.path.join(input_dir, filename)
        
        # Augment rotation
        augment_rotation(image_path, output_dir_rotation)
        
        # Augment brightness
        augment_brightness(image_path, output_dir_brightness)
        
        # Augment noise
        augment_noise(image_path, output_dir_noise)

```
